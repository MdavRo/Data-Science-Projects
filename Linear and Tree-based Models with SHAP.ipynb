# Introduction

Lung cancer is a type of cancer that begins in the lungs and is the leading cause of cancer deaths worldwide. In 2020, there were more than 2.2 million new cases of lung cancer worldwide. While smoking is the leading cause of lung cancer, it can also occur in people who have never smoked (WCRF International, 2022).

Lung cancer is broadly classified into two main types: small-cell lung cancer (SCLC) and non-small-cell lung cancer (NSCLC). SCLC, although less common than NSCLC, is almost exclusively found in heavy smokers. On the other hand, NSCLC has several subtypes, including adenocarcinoma, squamous cell carcinoma, and large cell carcinoma. Adenocarcinoma is the most common form of lung cancer in nonsmokers and younger people and starts in the mucus-secreting cells in the deeper part of the lungs. Squamous cell carcinoma and large cell carcinoma start in different types of cells in the lungs and have different growth patterns, spreading processes, and treatment approaches.

**Most people link lung cancer to smoking but its not the case all the time.** Various environmental and occupational factors contribute to lung cancer risk. Exposure to substances such as radon, air pollution, asbestos, diesel exhaust, and certain other chemicals can increase the likelihood of developing the disease. However, it is worth noting that a small portion of lung cancer cases occurs in individuals with no known risk factors, underscoring its complex nature. Further, passive inhalation of cigarette smoke, also known as secondhand smoke, has been linked to lung cancer in nonsmokers.

# **Machine Learning Implementation**

```python
import time
import warnings
from itertools import product
from collections import Counter
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.exceptions import ConvergenceWarning
from IPython.display import display, HTML

warnings.filterwarnings("ignore", category=FutureWarning)
# warnings.filterwarnings("ignore", category=RuntimeWarning)
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
def model_comparator(X, y, scaler=None, \
                     feature_name=None, target_name=None, trials=100, \
                     Cs=None, gammas=None, ratio=25):
    # scaler, feature_name, and target_name are optional parameters that can be used to provide
    # additional information for scaling, feature names, and target names, respectively
    # 'trials' specifies the number of trials or iterations to perform for each model
    # 'Cs' and 'gammas' are parameters for the models that require them
    # 'ratio' is a parameter used in the knn_trials function
    # The function performs comparisons for various machine learning models
    # by calling different trial functions for each model

    start_time = time.time() # record the start time
    knn = knn_trials(X, y, scaler=scaler, summarize=True,
                     trials=trials, ratio=ratio)
    print(f'knn Complete : {(time.time() - start_time):0.2f} secs.') # (time.time() - start_time)
                                                                     # indicates the time taken for each model

    start_time = time.time()
    log_l1 = logr_trials(X, y, trials=trials, scaler=scaler,
                      reg='l1', summarize=True, Cs=Cs,
                      feature_name=feature_name, target_name=target_name)
    print(f'log_l1 Complete : {(time.time() - start_time):0.2f} secs.')

    start_time = time.time()
    log_l2 = logr_trials(X, y, trials=trials, scaler=scaler,
                      reg='l2', summarize=True, Cs=Cs,
                      feature_name=feature_name, target_name=target_name)
    print(f'log_l2 Complete : {(time.time() - start_time):0.2f} secs.')

    start_time = time.time()
    lsvm_l1 = lsvm_trials(X, y, trials=trials, scaler=scaler,
                      reg='l1', summarize=True, Cs=Cs,
                      feature_name=feature_name, target_name=target_name)
    print(f'lsvm_l1 Complete : {(time.time() - start_time):0.2f} secs.')

    start_time = time.time()
    lsvm_l2 = lsvm_trials(X, y, trials=trials, scaler=scaler,
                    reg='l2', summarize=True, Cs=Cs,
                    feature_name=feature_name, target_name=target_name)
    print(f'lsvm_l2 Complete : {(time.time() - start_time):0.2f} secs.')

    start_time = time.time()
    nsvm_rbf = nsvm_trials(X, y, trials=trials, scaler=scaler, Cs=Cs,
                           summarize=True, gammas=gammas)
    print(f'nsvm_rbf Complete : {(time.time() - start_time):0.2f} secs.')

    cols = ['Machine Learning Method', 'Test Accuracy',
            'Best Parameter', 'Top Predictor Variable']
    df = pd.DataFrame([knn, log_l1, log_l2, lsvm_l1, lsvm_l2, nsvm_rbf],
                        columns=cols) # creates another dataframe using the performance metrics obtained from the trial functions

    display(HTML(df.to_html().replace('\\n', '<br>')))

def knn_trials(X, y, trials=100, scaler=None, \
            neighborhoods=None, ratio=0.25, graph=True,\
            summarize=True):
#     'scaler' is an optional parameter that allows scaling of the input features.
#     'neighborhoods' is an optional parameter that defines the range of values for the number of neighbors to consider.
#     'ratio' specifies the test size ratio when splitting the dataset into train and test sets.
#     'graph' is a boolean parameter that controls whether a graph of the accuracy is generated.
#     'summarize' is a boolean parameter that controls whether the function returns a summary of the results.

    seeds = np.arange(trials)

    # Defining the neighborhoods
    if neighborhoods is None:
        neighborhoods = np.arange(1, (len(X)//4) - 1, 1)
    else:
        neighborhoods = np.arange(1, neighborhoods + 1, 1)

    train_score = pd.DataFrame()
    test_score = pd.DataFrame()

    for col, seed in enumerate(seeds):
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                test_size=ratio,
                                                random_state=seed,
                                                stratify=y)

        if scaler is not None:
            try:
                scaler.fit(X_train)
                X_train = scaler.transform(X_train)
                X_test = scaler.transform(X_test)
            except BaseException as e:
                print(f'ERROR: Ensure that scaler is a Scaler class object')
                print(f'{e}')
                break

        # Result per seed/ trial
        train_accuracy = []
        test_accuracy = []
        for n in neighborhoods:
            knn = KNeighborsClassifier(n_neighbors=n)
            knn.fit(X_train, y_train)
            train_accuracy.append(knn.score(X_train, y_train))
            test_accuracy.append(knn.score(X_test, y_test))

        # Storing the result in the dataframe
        train_score[col] = train_accuracy
        test_score[col] = test_accuracy

    # Summarizing the results
    mean_train = train_score.mean(axis=1)
    mean_test = test_score.mean(axis=1)
    ideal_n = neighborhoods[np.argmax(mean_test.values)] # np.argmax() is a NumPy function that returns
                                                         # the index of the maximum value in an array


    if graph:
        # Define values to plot the error/ standard deviation
        train_err = (np.array([[-1],[1]]) *
                    (train_score.std(axis=1) + mean_train).values)
        test_err = (np.array([[-1],[1]]) *
                    (test_score.std(axis=1) + mean_test).values)

        # Plotting Paraphernalia
        plt.figure(figsize=(5,5))
        plt.plot(neighborhoods, mean_train, label='train score',
                color='blue', marker='o', linestyle='dashed')
        plt.plot(neighborhoods, mean_test, label='test score',
                color='green', marker='^', linestyle='dashed')
        plt.fill_between(neighborhoods, test_err[0], test_err[1],
                    alpha=0.2, color='green')
        plt.title(f'Mean kNN Accuracy graphs of the dataset')

        plt.title('Accuracy Behavior of kNN Classifier at varying '
                  f'n-Neighbors\ntrials={trials}, test_ratio={ratio},\n'
                  f'best_n={ideal_n}, max_score={mean_test.max():0.4f}')
        plt.xlabel('Neighbors, n')
        plt.ylim(min(mean_test)*3/4, 1)
        plt.ylabel('Accuracy/ Score')
        plt.legend(loc='lower left')
        plt.show()

    if summarize:
        return ['kNN', mean_test.max(), f'n_neighbor={ideal_n}', 'NA']
    else:
        return [mean_train, mean_test]

def logr_trials(X, y, trials=100, scaler=False, \
            ratio=0.25, reg='l2', graph=True, \
            Cs=None, summarize=True, feature_name=None,\
            target_name=None):

    # Seed Generation
    # seeds = []
    # while len(seeds) < trials:
    #     s = np.random.randint(0, trials*10)
    #     if s not in seeds: seeds.append(s)
    # seeds = np.array(seeds)
    seeds = np.arange(trials)

    # Defining the C Parameter
    if Cs is None:
        Cs = [1e-8, 1e-4, 1e-3, 1e-2, 0.1, 0.2, 0.4, 0.75, 1,
                  1.5, 3, 5, 10, 15,  20, 100, 300, 1000, 5000]

    # Defining solver
    if reg == 'l1':
        solver = 'liblinear' # lasso
    else:
        solver = 'lbfgs' # ridge

    train_score = pd.DataFrame()
    test_score = pd.DataFrame()
    weighted_coefs = pd.DataFrame()


    for col, seed in enumerate(seeds):
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                test_size=ratio,
                                                random_state=seed,
                                                stratify=y)

        if scaler is not None:
            try:
                scaler.fit(X_train)
                X_train = scaler.transform(X_train)
                X_test = scaler.transform(X_test)
            except BaseException as e:
                print(f'ERROR: Ensure that scaler is a Scaler class object')
                print(f'{e}')
                break

        training_accuracy = []
        test_accuracy = []
        coefs = []
        for C in Cs:

            logr = LogisticRegression(C=C, penalty=reg, solver=solver,
                                    max_iter=10000)
            logr.fit(X_train, y_train)
            training_accuracy.append(logr.score(X_train, y_train))
            test_accuracy.append(logr.score(X_test, y_test))
            coefs.append(logr.coef_)

        train_score[col] = training_accuracy
        test_score[col]= test_accuracy
        weighted_coefs[col] = coefs

    mean_train = train_score.mean(axis=1)
    mean_test = test_score.mean(axis=1)
    mean_coef = pd.Series(np.mean(weighted_coefs.values, axis=1))
    ind = np.argmax(mean_test.values)

    # Generating the list of Relevant Features
    top_ind = np.argmax(abs(mean_coef[ind]), axis=1)
    top_wts = abs(mean_coef[ind]).max(axis=1)


    # Dummy feature_name Generator
    if feature_name is None:
        feature_name = [f'feature{i}' for i in range(X.shape[1])]

    top_name = [feature_name[i] for i in top_ind]

    # Generating dummy target_name

    if target_name is None:
        target_name = [f'class{i}' for i in np.unique(y)]

    top_report = '\n'.join([f'{t}:{f} = {w:0.4f}'
                              for t, f, w in zip(target_name,
                                                 top_name,
                                                 top_wts)])

    if graph:
        # Define values to plot the error/ standard deviation
        train_err = (np.array([[-1],[1]]) *
                    (train_score.std(axis=1) + mean_train).values)
        test_err = (np.array([[-1],[1]]) *
                    (test_score.std(axis=1) + mean_test).values)

        # Plot particulars
        fig, ax = plt.subplots(figsize=(7,7))
        ax.set_xscale('log')
        plt.plot(Cs, mean_train, label="training accuracy",
                color='blue', marker='o', linestyle='dashed')
        plt.plot(Cs, mean_test, label="test accuracy",
                color='green', marker='^', linestyle='-')
        plt.fill_between(Cs, test_err[0], test_err[1],
                    alpha=0.2, color='green')

        plt.ylabel('Accuracy/ Score, $R^2$')
        plt.ylim(min(mean_test)*3/4, 1)
        plt.xlabel('Regularization parameter, $C$')
        plt.title(f'$R^2$ Behavior of {reg.upper()}-Regularized Logistics '
                   'Regression\nagainst the Regularization Parameter, $C$\n'
                  f'trials={trials}, test_ratio={ratio},\nmax'
                  f'$C$={Cs[ind]}, max$R^2$={mean_test.max():0.4f}')
        plt.legend(loc='lower left')

    if summarize:
        return [f'Logistic Regression with {reg.upper()}-Regularization',
                mean_test.max(),
                f'C = {Cs[ind]}', top_report]
    else:
        return [mean_train, mean_test, mean_coef]



def lsvm_trials(X, y, trials=100, scaler=False, graph=True, \
            ratio=0.25, reg='l2', Cs=None, \
            summarize=True, feature_name=None, \
            target_name=None):

    # Seed Generation
    # seeds = []
    # while len(seeds) < trials:
    #     s = np.random.randint(0, trials*10)
    #     if s not in seeds: seeds.append(s)
    # seeds = np.array(seeds)
    seeds = np.arange(trials)

    # Defining the C Parameter
    if Cs is None:
        Cs = [1e-8, 1e-4, 1e-3, 1e-2, 0.1, 0.2, 0.4, 0.75, 1,
                  1.5, 3, 5, 10, 15,  20, 100, 300, 1000, 5000]


    train_score = pd.DataFrame()
    test_score = pd.DataFrame()
    weighted_coefs = pd.DataFrame()

    for col, seed in enumerate(seeds):
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                test_size=ratio,
                                                random_state=seed,
                                                stratify=y)

        if scaler is not None:
            try:
                scaler.fit(X_train)
                X_train = scaler.transform(X_train)
                X_test = scaler.transform(X_test)
            except BaseException as e:
                print(f'ERROR: Ensure that scaler is a Scaler class object')
                print(f'{e}')
                break

        training_accuracy = []
        test_accuracy = []
        coefs = []

        for C in Cs:

            lsvm = LinearSVC(C=C, penalty=reg, dual=False, max_iter = 1000000)
            lsvm.fit(X_train, y_train)
            training_accuracy.append(lsvm.score(X_train, y_train))
            test_accuracy.append(lsvm.score(X_test, y_test))
            coefs.append(lsvm.coef_)

        train_score[col] = training_accuracy
        test_score[col]= test_accuracy
        weighted_coefs[col] = coefs

    mean_train = train_score.mean(axis=1)
    mean_test = test_score.mean(axis=1)
    mean_coef = pd.Series(np.mean(weighted_coefs.values, axis=1))

    ind = np.argmax(mean_test.values)

    # Generating the list of Relevant Features
    top_ind = np.argmax(abs(mean_coef[ind]), axis=1)
    top_wts = abs(mean_coef[ind]).max(axis=1)

    # Dummy feature_name Generator
    if feature_name is None:
        feature_name = [f'feature{i}' for i in range(X.shape[1])]

    top_name = [feature_name[i] for i in top_ind]

    # Generating dummy target_name

    if target_name is None:
        target_name = [f'class{i}' for i in np.unique(y)]

    top_report = '\n'.join([f'{t}:{f} = {w:0.4f}'
                              for t, f, w in zip(target_name,
                                                 top_name,
                                                 top_wts)])

    if graph:
        # Define values to plot the error/ standard deviation
        train_err = (np.array([[-1],[1]]) *
                    (train_score.std(axis=1) + mean_train).values)
        test_err = (np.array([[-1],[1]]) *
                    (test_score.std(axis=1) + mean_test).values)

        # Plot particulars
        fig, ax = plt.subplots(figsize=(7,7))
        ax.set_xscale('log')
        plt.plot(Cs, mean_train, label="training accuracy",
                color='blue', marker='o', linestyle='dashed')
        plt.plot(Cs, mean_test, label="test accuracy",
                color='green', marker='^', linestyle='-')
        plt.fill_between(Cs, test_err[0], test_err[1],
                    alpha=0.2, color='green')
        plt.ylabel('Accuracy/ Score, $R^2$')
        plt.ylim(min(mean_test)*3/4, 1)
        plt.xlabel('Regularization parameter, $C$')

        plt.title(f'$R^2$ Behavior of {reg.upper()}-Regularized LinearSVM '
                   '\nagainst the Regularization Parameter, $C$\n'
                  f'trials={trials}, test_ratio={ratio},\nmax'
                  f'$C$={Cs[ind]}, max$R^2$={mean_test.max():0.4f}')
        plt.legend(loc='lower left')

    if summarize:
        return [f'Linear SVM with {reg.upper()}-Regularization',
                mean_test.max(),
                f'C = {Cs[ind]}', top_report]
    else:
        return [mean_train, mean_test, mean_coef]


def nsvm_trials(X, y, trials=100, scaler=False, graph=True, \
            ratio=0.25, Cs=None, gammas=None,\
            summarize=True):

    # Seed Generation
    # seeds = []
    # while len(seeds) < trials:
    #     s = np.random.randint(0, trials*10)
    #     if s not in seeds: seeds.append(s)
    # seeds = np.array(seeds)
    seeds = np.arange(trials)

    # Defining the C Parameter
    if Cs is None:
        Cs = [1e-8, 1e-4, 1e-3, 1e-2, 0.1, 0.2, 0.4, 0.75, 1,
              1.5, 3, 5, 10, 15,  20, 100, 300, 1000, 5000]


    # Defining the gamma Parameter
    if gammas is None:
        gammas = [1e-8, 1e-4, 1e-3, 1e-2, 0.1, 0.2, 0.4, 0.75, 1,
                  1.5, 3, 5, 10, 15,  20, 100, 300, 1000, 5000]


    train_score = (pd.DataFrame(list(product(Cs, gammas)),
                                columns=['C', 'gamma'])
                     .set_index(['C', 'gamma']))
    test_score = (pd.DataFrame(list(product(Cs, gammas)),
                               columns=['C', 'gamma'])
                     .set_index(['C', 'gamma']))

    # weighted_coefs = pd.DataFrame()

    for col, seed in enumerate(seeds):
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                test_size=ratio,
                                                random_state=seed,
                                                stratify=y)

        if scaler is not None:
            try:
                scaler.fit(X_train)
                X_train = scaler.transform(X_train)
                X_test = scaler.transform(X_test)
            except BaseException as e:
                print(f'ERROR: Ensure that scaler is a Scaler class object')
                print(f'{e}')
                break

        training_accuracy = []
        test_accuracy = []
        # coefs = []

        for C, gamma in product(Cs, gammas):

            nsvm = SVC(kernel='rbf', gamma=gamma, C=C, max_iter = 1000000)
            nsvm.fit(X_train, y_train)
            training_accuracy.append(nsvm.score(X_train, y_train))
            test_accuracy.append(nsvm.score(X_test, y_test))
            # coefs.append(nsvm.coef_)


        train_score[col] = training_accuracy
        test_score[col]= test_accuracy
        # rows are defined by the sequence of c_gamma

        # weighted_coefs[col] = coefs

    mean_train = train_score.mean(axis=1)
    mean_test = test_score.mean(axis=1)
    # mean_coef = pd.Series(np.mean(weighted_coefs.values, axis=1))

    ind = np.argmax(mean_test.values)
    # Best parameters
    C, g = mean_test.index[ind]



    if graph:

        # Extract information for plotting

        gamma_X = mean_test.loc[(C)].index.to_list()
        mean_test_Y = mean_test.loc[(C)].values
        mean_train_Y = mean_train.loc[(C)].values

        # Define values to plot the error/ standard deviation
        train_err = (np.array([[-1],[1]]) *
                    (train_score.std(axis=1).loc[(C)] +
                     mean_train.loc[(C)]).values)
        test_err = (np.array([[-1],[1]]) *
                    (test_score.std(axis=1).loc[(C)] +
                     mean_test.loc[(C)]).values)

        # Plot particulars
        fig, ax = plt.subplots(figsize=(7,7))
        ax.set_xscale('log')
        plt.plot(gamma_X, mean_train_Y, label="training accuracy",
                color='blue', marker='o', linestyle='dashed')
        plt.plot(gamma_X, mean_test_Y, label="test accuracy",
                color='green', marker='^', linestyle='-')
        plt.fill_between(gamma_X, test_err[0], test_err[1],
                    alpha=0.2, color='green')
        plt.ylabel('Accuracy/ Score, $R^2$')
        plt.ylim(min(mean_test)*3/4, 1)
        plt.xlabel('Radial parameter, $\gamma$')
        plt.title(f'$R^2$ Behavior of NonlinearSVM (Radial Based Function)'
                   '\nagainst the Radial Parameter, $\gamma$\n'
                  f'trials={trials}, test_ratio={ratio}, max_C = {C}\nmax_'
                  f'$\gamma$={g}, max_$R^2$={mean_test.max():0.4f}')
        plt.legend(loc='lower left')

    if summarize:
        return [f'Nonlinear SVM (Radial Basis Function)', mean_test.max(),
                f'gamma = {g}\n C = {C}', 'NA']
    else:
        return [mean_train, mean_test]

def PCC_calculator(target, target_name=[], dataset_name='given', graph=True):
    """Returns the Proportion Chance Criteria (PCC) of a dataset's
    target labels.

    Parameter
    ---------
    target : 1-D np.array or  pd.Series or list
        Countable list of targets. This will be passed through the Counter
    function.

    target_name : list of strings
        Numerical or non-numerical labels for the targets to be used as labels
    of the xtickmarks of the graph

    dataset_name : string
        Name of the dataset to be used in the graph/ visualization

    graph : Bool
        If `True` a frequency graph of the targets will be shown

    Return
    ------
    PCC : Float
        1.25PCC value of the target
    """
    target_count = Counter(target)

    df_target_count = pd.DataFrame.from_dict(target_count,
                                            orient='index',
                                            columns=['n_i'])

    PCC = ((df_target_count.n_i/ df_target_count.n_i.sum())**2).sum()*1.25*100

    if not any(target_name):
        target_name = df_target_count.index.to_list()

    if graph:
        ax = df_target_count.plot(kind='bar', rot=0, figsize=(7,7))
        ax.get_legend().remove()
        ax.set_ylabel('Frequency of each target')
        ax.set_xlabel(f'Target labels of the {dataset_name} Dataset')
        ax.set_title(f'"Targets" Distribution of the {dataset_name} Dataset\n'
                     f'(1.25PCC = {PCC:0.2f}%)')
        ax.set_xticklabels(target_name)

    return PCC
```

**Findings:**

The values on the diagonal of the confusion matrix represent the number of correctly classified instances for each class, while the values off the diagonal represent misclassifications. So, the **correct classification rate or accuracy** is the sum of the number on the diagonal divided by the sample size in the test data. In our example, that is (91 + 76 + 83)/250 = 100%.

**Precision** measures the proportion of correctly predicted positive instances (true positives) out of all instances predicted as positive for a specific class.
Precision = (True Positives) / (True Positives + False Positives)

**Sensitivity (or Recall)** measures the proportion of correctly predicted positive instances (true positives) out of all actual positive instances for a specific class.
Recall = (True Positives) / (True Positives + False Negatives)

**F1-score** is the harmonic mean of precision and recall. It provides a balanced measure that considers both precision and recall for a specific class.
F1-score = 2 * (Precision * Recall) / (Precision + Recall)
